# Step 1: Import libraries

```python
from transformers import pipeline
```

# Step 2: Load pre-trained model

```python
classifier = pipeline("sentiment-analysis")
```

# Step 3: Define stress scoring function

```python
def text_stress_analysis(text):
    result = classifier(text)[0]
    label = result['label']
    score = result['score']

    # Map sentiment to stress (basic rule)
    # POSITIVE → Low Stress, NEUTRAL → Medium, NEGATIVE → High Stress
    if label == 'POSITIVE':
        stress_level = 'Low Stress'
        stress_score = 30 * (1 - score)  # lower probability → higher stress
    elif label == 'NEUTRAL':
        stress_level = 'Medium Stress'
        stress_score = 50 + 20*(1 - score)
    else:  # NEGATIVE
        stress_level = 'High Stress'
        stress_score = 70 + 30*score

    stress_score = min(100, stress_score)  # cap at 100
    return {'stress_score': stress_score, 'stress_level': stress_level, 'text_label': label}
```

# Step 4: Test the model

```python
text_input = "I am feeling very overwhelmed and anxious about my work deadlines."
result = text_stress_analysis(text_input)

print("Text Input:", text_input)
print("Predicted Stress Score:", result['stress_score'])
print("Predicted Stress Level:", result['stress_level'])
print("Underlying Sentiment Label:", result['text_label'])
```
